# 35. Search Insert Position
從中間開始查找,讓時間複雜度為O(logn)

```python
nums = [1,3,4,5,6,7,8]
target = 2
```

```python
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:

        def helper(nums: List[int], start: int) -> int:
            if len(nums)==0:
                return start
            
            mid_index = len(nums)//2
            mid = nums[mid_index]

            if mid == target:
                return start + mid_index
            elif mid > target:
                return helper(nums[:mid_index], start)
            else:
                return helper(nums[mid_index+1:], start + mid_index + 1)

        return helper(nums,0)
    
# recursive version
```

```python
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:

        left , right = 0 , len(nums)-1

        while left <= right:
            mid = (left+right)//2

            if nums[mid] == target:
                return mid
            elif nums[mid] > target:
                right = mid - 1
            else:
                left = mid + 1

        #  left指向大於或等於 target 的最小值的位置
        return left 
    
# iterative version
```

```python
Solution().searchInsert(nums,target)
```

# 74. Search a 2D Matrix

```python
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        
        left , right = 0 , len(matrix[0])-1
        up , down = 0 ,  len(matrix)-1

        while up <= down:
            mid_m = (up+down)//2
            point_m = matrix[mid_m][left]

            if point_m < target:
                up = mid_m+1
            elif point_m > target:
                down = mid_m-1
            else:
                return True
            
        
        while left <= right:
            mid_n = (left+right)//2
            point_n = matrix[down][mid_n]

            if point_n < target:
                left = mid_n+1
            elif point_n > target:
                right = mid_n-1
            else:
                return True

        return False
        
```

# 162. Find Peak Element

```python
nums = [1,2,1,3,5,6,4]
#Output: 5
```

```python
class Solution:
    def findPeakElement(self, nums: List[int]) -> int:
        
        temp = float("-inf")
        ans_index = -1
        for index,i in enumerate(nums):
            if i>temp:
                temp = i
                ans_index = index

        return ans_index

# for loop
```

```python
class Solution:
    def findPeakElement(self, nums: List[int]) -> int:
        
        left , right = 0 , len(nums)-1

        while left < right:
            mid = (left+right)//2

            if nums[mid] < nums[mid+1]:
                left =  mid + 1
            elif nums[mid] > nums[mid+1]:
                right = mid

        return left
    
# binary_search
```

# 34. Find First and Last Position of Element in Sorted Array

```python
nums = [5,7,7,8,8,10]
target = 7
# Output: [3,4]
```

```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        
        left , right = 0 , len(nums)-1

        while left <= right:
            mid = (left+right)//2

            if nums[mid] == target:
                r = mid 
                while r+1 < len(nums):
                    if nums[r+1] == target:
                        r+=1
                    else:
                        break
                
                l = mid
                while l-1 >= 0:
                    if nums[l-1] == target:
                        l-=1
                    else:
                        break

                return [l,r]

            elif nums[mid] > target:
                right = mid - 1
            else:
                left = mid +1

        return [-1,-1]
```

```python
Solution().searchRange(nums,target)
```

# 153. Find Minimum in Rotated Sorted Array

```python
# nums = [2,4,5,6,7,0,1]
# nums = [0,1,2,4,5,6,7]
nums = [6,7,0,1,2,4,5]
# nums = [5,6,7,0,1,2,4]
```

```python
class Solution:
    def findMin(self, nums: List[int]) -> int:

        left , right = 0 , len(nums)-1
        left_num = nums[left]

        while left < right:
            mid = (left+right)//2

            if mid+1 < len(nums) or mid-1 >= 0:
                if nums[mid] > nums[mid+1]:
                    return nums[mid+1]
                elif nums[mid] < nums[mid-1]:
                    return nums[mid]
                elif nums[mid] < nums[mid+1] and nums[mid]>left_num:
                    left = mid+1
                else:
                    right = mid-1

        return min(left_num,nums[left])
    
# 有過但程式碼太冗長
```

```python
class Solution:
    def findMin(self, nums: List[int]) -> int:

        left , right = 0 , len(nums)-1
        while left < right:

            mid = (left+right)//2
    
            if nums[mid]>=nums[left]:
                if nums[left] > nums[right]:  # 表示最小值在右半部分
                    left = mid + 1
                else:  # 表示數組本身有序，最小值在左半部分
                    right = mid
            else:
                right = mid

        return nums[left]
```
